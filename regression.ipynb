```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Import Dataset

```python
df = pd.read_csv('./data/Salary_Data.csv')
df.head()
```

```python
X = df.iloc[:, 0].values
y = df.iloc[:, -1].values
```

```python
X
```

```python
y
```

## Split the dataset into Train and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

## Train with Linear Regression

```python
from sklearn.linear_model import LinearRegression

regression = LinearRegression()

regression.fit(X_train.reshape(-1, 1), y_train.reshape(-1, 1))
```

```python
y_pred = regression.predict(X=X_test.reshape(-1, 1))
y_pred
```

## Visualize

```python
plt.scatter(X_test, y_test)
plt.plot(X_test, regression.predict(X_test.reshape(-1, 1)), color='red')
plt.title('Salary vs Experience (Test set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
```

```python
plt.scatter(X_train, y_train)

y_train_pred = regression.predict(X=X_train.reshape(-1, 1))
plt.plot(X_train, y_train_pred, color='red')

plt.title('Salary vs Experience (Train set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```
